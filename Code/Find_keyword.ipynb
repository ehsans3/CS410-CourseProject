```python
import nltk
from operator import itemgetter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
stopwords = stopwords.words('english')

print(stopwords)
```

```python
with open(r"C:\CS410-CourseProject\Topics Output\topic_cs410.csv") as text_input, open("C:\\CS410-CourseProject\\Topics Output\\topic_cs410.txt", 'w') as text_output:
    content = text_input.read()
    content = content.replace(',', '')
    content = content.replace('topic_name', '')
    content = content.replace('This', '')
    content = content.replace('It', '')
    content = content.replace('The', '')
    content = content.replace('discusses', ' ')
    content = content.replace('used', ' ')
    content = content.replace('use', ' ')
    content = content.replace('using', ' ')
    content = content.replace('different', ' ')
    content = content.replace('covers', ' ')
    content = content.replace('"', ' ')
    text_output.write(content)
```

```python
file_filter = open(r"C:\CS410-CourseProject\Topics Output\topic_cs410.txt") 
line = file_filter.read()
words = line.split() 
for word in words: 
    if not word in stopwords: 
        appendFile = open("C:\\CS410-CourseProject\\Topics Output\\filter_topic_cs410.txt", 'a') 
        appendFile.write(" "+ word) 
        appendFile.close() 
```

```python
number_of_words = 0
counts = {}
p = {}
with open(r"C:\CS410-CourseProject\Topics Output\filter_topic_cs410.txt") as input_file:
    for line in input_file:
        for word in line.split():
            word = word.strip(".,!?:;'\"")
            word = word.lower()
            if not word.isnumeric(): 
                number_of_words = number_of_words + 1
            if word not in counts:
                counts[word] = 0
            counts[word] = (counts[word] + 1)

with open("C:\CS410-CourseProject\Topics Output\count_topic_cs410.txt",'w') as count_outfile:
    for word in sorted(counts, key = counts.get , reverse = True):
        print(word, counts[word])
        count_outfile.write("{}: {}\n".format(word, counts[word]))
```

```python
with open("C:\CS410-CourseProject\Topics Output\p_topic_cs410.txt",'w') as p_outfile:
    for word in sorted(counts, key = counts.get , reverse = True):
        p[word] = round ((counts[word] / number_of_words), 6)
        print(word, p[word])
        p_outfile.write("{}: {}\n".format(word, p[word]))
```

```python
N = 30
top_30 = dict(sorted(counts.items(), key = itemgetter(1), reverse=True)[:N])
 
print(str(top_30))
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(top_30)), top_30.values(), align = 'center')
plt.xticks(rotation = 90)
plt.xticks(range(len(top_30)), list(top_30.keys()))

plt.show()
```
